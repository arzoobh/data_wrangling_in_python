```python
import os
import pprint
import csv
```

```python
DATADIR = ""
DATAFILE = 'beatles-discography.csv'
```

```python
#parse file function
def parse_csv(datafile):
    data =[]
    n = 0
    with open(datafile, 'r') as f:
        r = csv.DictReader(f)
        for line in r:
            data.append(line)
    return(data)
```

```python
if __name__ == '__main__':
    datafile = os.path.join(DATADIR, DATAFILE)
    parse_csv(datafile)
    d = parse_csv(datafile)
    pprint.pprint(d)
```
